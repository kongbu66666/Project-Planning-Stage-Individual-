```r
# ============================
# Setup: load packages & data
# ============================
library(tidyverse)

players  <- read_csv("players.csv")
sessions <- read_csv("sessions.csv")

# Inspect structure
glimpse(players)
glimpse(sessions)

# ============================
# (1) Data description helpers
# ============================

# Basic sizes (you can report these numbers in the Data Description section)
n_players       <- nrow(players)
n_vars_players  <- ncol(players)
n_sessions      <- nrow(sessions)
n_vars_sessions <- ncol(sessions)

n_players; n_vars_players
n_sessions; n_vars_sessions

# Simple summaries
summary(players)
summary(sessions)

# ============================
# Minimal wrangling for players
# ============================
players <- players |>
  mutate(
    subscribe  = if_else(subscribe, "Subscribed", "Not subscribed"),
    subscribe  = factor(subscribe),
    experience = factor(experience),
    gender     = factor(gender)
  )

glimpse(players)

# ============================
# (3) Means of numeric variables in players.csv
# ============================
players_means <- players |>
  select(where(is.numeric)) |>
  summarise(across(
    everything(),
    ~ round(mean(.x, na.rm = TRUE), 2)
  )) |>
  pivot_longer(
    everything(),
    names_to  = "variable",
    values_to = "mean_value"
  )

players_means   # This table can be shown in the report

# ============================
# (3) Exploratory plots
# ============================

# 1. Counts of players by subscription status
players |>
  ggplot(aes(x = subscribe)) +
  geom_bar() +
  labs(
    x     = "Newsletter subscription status",
    y     = "Number of players",
    title = "Counts of players by newsletter subscription"
  ) +
  theme_minimal()

# 2. Total played hours by subscription status
players |>
  ggplot(aes(x = subscribe, y = played_hours)) +
  geom_boxplot() +
  labs(
    x     = "Newsletter subscription status",
    y     = "Total played hours",
    title = "Played hours by newsletter subscription"
  ) +
  theme_minimal()

# 3. Age by subscription status
players |>
  ggplot(aes(x = subscribe, y = Age)) +
  geom_boxplot() +
  labs(
    x     = "Newsletter subscription status",
    y     = "Age (years)",
    title = "Age distribution by newsletter subscription"
  ) +
  theme_minimal()

# 4. Counts by experience level (optional)
experience_counts <- players |>
  count(experience, sort = TRUE)

experience_counts
```

On average, players in this sample have spent about 5.85 hours on the Minecraft research server and are around 21 years old. This suggests that the dataset mainly consists of young adult players with relatively low to moderate total play time, but there are also some players with much higher engagement levels, as seen in the later plots.


## (3) Exploratory Data Analysis and Visualization

### Counts of players by newsletter subscription

The bar plot of newsletter subscription status shows a clear class imbalance: there are many more **subscribed** players than **not subscribed** players (144 vs. 52). This imbalance is important for later modelling, because some classification methods may favour the majority class if no additional steps are taken to handle it.

### Played hours by newsletter subscription

The boxplot of total played hours by subscription status indicates that subscribed players tend to have **higher played hours** than non-subscribed players. The median played hours for subscribed players is slightly higher, and there are many more extreme high values (outliers) in the subscribed group. This pattern suggests that players who are more engaged with the server are also more likely to subscribe to the newsletter, so played hours is likely to be a useful predictor.

### Age distribution by newsletter subscription

The boxplot of age by subscription status shows that the age distributions of subscribed and non-subscribed players are quite similar. Both groups have medians around the early twenties and a similar spread, with a few older outliers. Any differences in age between the two groups appear to be small, so age may be a weaker predictor of newsletter subscription compared with played hours.


## (4) Methods and Plan

To answer my question, I plan to use a classification model to predict newsletter subscription from player characteristics. My primary method will be **logistic regression**, with `subscribe` as the binary response and predictors such as `played_hours`, `Age`, `experience`, and `gender`.

Logistic regression is appropriate because the outcome is binary and the model provides interpretable coefficients and predicted probabilities. The main assumptions are that observations (players) are independent, the log-odds of subscribing are approximately a linear function of the predictors, and there is no extreme multicollinearity. A limitation is that complex non-linear relationships may not be fully captured.

I will randomly split the data into a **training set (75%)** and a **test set (25%)** at the player level. On the training set, I will use **cross-validation** to fit and assess logistic regression (and possibly compare with another classifier such as k-nearest neighbours). Final performance will be evaluated on the held-out test set using appropriate classification metrics.
